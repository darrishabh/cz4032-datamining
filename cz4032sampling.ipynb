```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#getting the business dataset
business_dataset = pd.read_csv('/Users/rishabaryan/Desktop/CZ4032/githubclone/cz4032-datamining/yelp_academic_dataset_business.csv')
```

```python
#getting the review dataset
review_dataset = pd.read_csv('/Users/rishabaryan/Desktop/CZ4032/githubclone/cz4032-datamining/yelp_academic_dataset_review.csv')
```

```python
#getting the user dataset
user_dataset = pd.read_csv('/Users/rishabaryan/Desktop/CZ4032/githubclone/cz4032-datamining/yelp_academic_dataset_user.csv')
```

```python
#List of business IDs with Restaurant/Food or both:

list_of_restos = []
for i in range(len(business_dataset)):
    if type(business_dataset.loc[i, "categories"]) == str:
            thisi = business_dataset.loc[i, "categories"].split(", ")
            if (set(["Restaurants","Food"]) & set(thisi)):
                list_of_restos.append(business_dataset.loc[i, "business_id"])
                

list_of_restos
```

```python
#filtering restaurant reviews
resto_reviews = review_dataset.loc[review_dataset['business_id'].isin(list_of_restos)]
```

```python
#selecting required columns
user_dataset_main = user_dataset[['user_id','fans','useful']]
```

```python
#normalizing the dataset
user_dataset_main['fans'] = user_dataset_main['fans']/max(user_dataset_main['fans'])
user_dataset_main['useful'] = user_dataset_main['useful']/max(user_dataset_main['useful'])
```

```python
#creating a score column
user_dataset_main['score'] = 0.5*user_dataset_main['fans'] + 0.5*user_dataset_main['useful']
```

```python
#sorting user dataset by score
user_dataset_sorted = user_dataset_main.sort_values(by = 'score', ascending = False)
```

```python
#choosing top 20 - LIST OF INFLUENCERS
user_dataset_top20 = user_dataset_sorted.head(20)
```

```python
#reviews of top 20 users
user_review_top20 = resto_reviews.loc[resto_reviews['user_id'].isin(user_dataset_top20['user_id'])]
```

```python
#restos of top 20 users
top20_user_restos = resto_reviews.loc[resto_reviews['business_id'].isin(user_review_top20['business_id'].unique())]
```

```python
top20_user_restos.head()
```

```python
influencer_reviews = top20_user_restos.loc[top20_user_restos['user_id'].isin(user_dataset_top20['user_id'])]
notinfluencer_reviews = top20_user_restos.loc[~top20_user_restos['user_id'].isin(user_dataset_top20['user_id'])]
influencer_reviews['influencer'] = 1
notinfluencer_reviews['influencer'] = 0
finaldataset = pd.concat([influencer_reviews , notinfluencer_reviews])
finaldataset
```

```python
finaldataset.to_csv('/Users/rishabaryan/Desktop/CZ4032/finaldataset.csv')
user_dataset_top20.to_csv('/Users/rishabaryan/Desktop/CZ4032/user_influencers.csv')
list_of_restos = pd.DataFrame(list_of_restos)
list_of_restos.to_csv('/Users/rishabaryan/Desktop/CZ4032/list_of_restos.csv')
```
